# 1. Introduction to Distance Metrics

Banyak algoritma Machine Learning bekerja berdasarkan konsep jarak antar data.

Ide utamanya:

Data yang mirip akan memiliki jarak yang dekat.

Data yang berbeda akan memiliki jarak yang jauh.

Distance metrics digunakan untuk:

- KNN
- K-Means
- Hierarchical Clustering
- DBSCAN
- Recommendation Systems



# Why Distance Metrics Matter

Misalkan terdapat dua mahasiswa:

Student A = [170 cm, 65 kg]

Student B = [172 cm, 67 kg]

Jarak mereka kecil sehingga dianggap mirip.

Distance metrics membantu algoritma menentukan tingkat kemiripan antar sampel.



# 2. Euclidean Distance

Euclidean Distance adalah distance metric paling umum.

Merupakan jarak garis lurus antara dua titik.


```

from sklearn.metrics.pairwise import euclidean_distances
import numpy as np

A = np.array([[1,2]])
B = np.array([[4,6]])

distance = euclidean_distances(A,B)

print(distance)
```


## Formula Euclidean Distance

d = sqrt((x1-x2)^2 + (y1-y2)^2)

Semakin kecil nilai d, semakin mirip kedua titik.



# 3. Manhattan Distance

Manhattan Distance menghitung jarak berdasarkan pergerakan horizontal dan vertikal.


```

from sklearn.metrics.pairwise import manhattan_distances

distance = manhattan_distances(A,B)

print(distance)
```


### Karakteristik Manhattan Distance

- Cocok untuk data grid
- Lebih tahan terhadap outlier dibanding Euclidean
- Sering digunakan pada high-dimensional data



# 4. Minkowski Distance

Minkowski merupakan generalisasi dari Euclidean dan Manhattan Distance.

Jika:

p=1 → Manhattan

p=2 → Euclidean



# Distance Metrics Overview

# 5. Understanding KNN

K-Nearest Neighbors (KNN) merupakan algoritma supervised learning.

Prinsip:

Sebuah data baru akan diklasifikasikan berdasarkan mayoritas tetangganya.


```

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    iris.data,
    iris.target,
    random_state=42
)

knn = KNeighborsClassifier(n_neighbors=3)

knn.fit(X_train, y_train)

print(knn.score(X_test, y_test))
```


## Code Explanation

n_neighbors=3

Model melihat 3 tetangga terdekat.

Prediksi dilakukan berdasarkan voting mayoritas.



# How KNN Works

Langkah-langkah:

1. Hitung jarak ke seluruh data training
2. Ambil K tetangga terdekat
3. Lakukan voting
4. Tentukan kelas akhir



# Choosing K

Nilai K sangat penting.

K terlalu kecil:

- Overfitting

K terlalu besar:

- Underfitting


```

for k in [1,3,5,7,9]:
    model = KNeighborsClassifier(n_neighbors=k)
    model.fit(X_train, y_train)
    print(k, model.score(X_test,y_test))
```


Eksperimen ini membantu memahami pengaruh nilai K terhadap performa model.



# 6. Hyperparameter Tuning in KNN

Hyperparameter utama:

- n_neighbors
- weights
- metric


```

from sklearn.model_selection import GridSearchCV

param_grid = {
    "n_neighbors":[1,3,5,7,9,11],
    "weights":["uniform","distance"]
}

grid = GridSearchCV(
    KNeighborsClassifier(),
    param_grid,
    cv=5
)

grid.fit(X_train,y_train)

print(grid.best_params_)
print(grid.best_score_)
```


## Why Hyperparameter Tuning?

Parameter optimal berbeda untuk setiap dataset.

Grid Search membantu menemukan kombinasi terbaik secara otomatis.



# Understanding Weights

uniform

Semua tetangga memiliki bobot sama.

distance

Tetangga yang lebih dekat memiliki pengaruh lebih besar.



# 7. Evaluating KNN Performance

Evaluasi dilakukan menggunakan:

- Accuracy
- Precision
- Recall
- F1 Score
- Confusion Matrix


```

from sklearn.metrics import classification_report

pred = knn.predict(X_test)

print(classification_report(y_test,pred))
```


## Classification Report

Menampilkan:

- Precision
- Recall
- F1 Score
- Support


```

from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,pred))
```


Confusion Matrix membantu memahami kesalahan klasifikasi yang dilakukan model.

# Importance of Feature Scaling

Karena menggunakan jarak, KNN sangat sensitif terhadap skala fitur.


```

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

knn_scaled = KNeighborsClassifier(n_neighbors=3)
knn_scaled.fit(X_train_scaled, y_train)

print(knn_scaled.score(X_test_scaled,y_test))
```


Biasanya scaling meningkatkan performa KNN secara signifikan.
