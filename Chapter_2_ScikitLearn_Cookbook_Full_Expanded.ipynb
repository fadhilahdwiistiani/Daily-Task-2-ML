# Introduction

Data preprocessing is one of the most critical stages in Machine Learning.

A powerful algorithm cannot compensate for poor-quality data.

The famous principle:

**Garbage In, Garbage Out (GIGO)**

states that poor input data will produce poor model performance.



# 1. The Impact of Raw Data on Model Performance

Machine learning algorithms learn patterns directly from data.

Problems such as:

- Missing values
- Outliers
- Noise
- Data leakage
- Inconsistent formatting

can significantly reduce model accuracy.



## Why Data Quality Matters

Poor data quality may cause:

1. Reduced accuracy
2. Overfitting
3. Underfitting
4. Bias
5. Unstable predictions

In real-world projects, data preparation often consumes more time than model building.



# 2. Common Data Issues

The chapter highlights several common problems:

### Missing Values
Some records contain incomplete information.

Extreme observations that differ greatly from the rest of the dataset.

### Categorical Variables
Machine learning models generally require numerical inputs.

### Different Feature Scales

# 3. Handling Missing Data

Missing data is one of the most common problems in data science.

Common strategies:

- Remove rows
- Remove columns
- Mean imputation
- Median imputation
- Most frequent imputation
- KNN imputation


```

import pandas as pd
import numpy as np

data = {
    'Age':[21,22,np.nan,24,25],
    'Score':[90,np.nan,80,70,85]
}

df = pd.DataFrame(data)

df
```


## Example Dataset

The dataset contains missing values represented by NaN.

Most scikit-learn models cannot directly train using missing values.


```

from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')

filled = imputer.fit_transform(df)

print(filled)
```


## Code Explanation

SimpleImputer replaces missing values.

Strategies:

- mean
- median
- most_frequent
- constant

The transformer first learns statistics using fit() and then fills values using transform().


```

from sklearn.impute import KNNImputer

knn = KNNImputer(n_neighbors=2)

filled = knn.fit_transform(df)

print(filled)
```


## KNN Imputation

Instead of using averages, KNNImputer estimates missing values from nearby observations.

Advantages:

- Often more accurate
- Preserves local structure

Disadvantages:

- More computationally expensive



# 4. Scaling Techniques

Many algorithms depend on distance calculations.

Examples:

- KNN
- SVM
- K-Means
- PCA

Scaling ensures all features contribute fairly.


```

from sklearn.preprocessing import StandardScaler
import numpy as np

X = np.array([
    [1000, 1],
    [2000, 2],
    [3000, 3]
])

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled)
```


## StandardScaler

Formula:

z = (x - mean) / std

Results:

- Mean becomes 0
- Standard deviation becomes 1

Widely used for linear models, SVMs, PCA, and neural networks.


```

from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()

scaled = scaler.fit_transform(X)

print(scaled)
```


## MinMaxScaler

Formula:

(x - min) / (max - min)

Output range:

0 → 1

Useful when preserving original distribution shape.


```

from sklearn.preprocessing import RobustScaler

scaler = RobustScaler()

scaled = scaler.fit_transform(X)

print(scaled)
```


## RobustScaler

Uses:

- Median
- Interquartile Range (IQR)

Advantages:

More resistant to outliers than StandardScaler.



# Scaling Comparison

| Method | Sensitive to Outliers |
|----------|----------|
| StandardScaler | Yes |
| MinMaxScaler | Yes |
| RobustScaler | No |



# 5. Encoding Categorical Variables

Machine learning models cannot directly understand text labels.

Example:

Color:

- Red
- Blue
- Green

must be converted into numbers.


```

from sklearn.preprocessing import LabelEncoder

labels = ['Red','Blue','Green','Red']

encoder = LabelEncoder()

encoded = encoder.fit_transform(labels)

print(encoded)
```


## Label Encoding

Example:

Red → 2

Blue → 0

Green → 1

Warning:

LabelEncoder introduces ordering relationships that may not actually exist.


```

from sklearn.preprocessing import OneHotEncoder
import pandas as pd

df = pd.DataFrame({
    'Color':['Red','Blue','Green']
})

encoder = OneHotEncoder(sparse_output=False)

encoded = encoder.fit_transform(df)

print(encoded)
```


## One-Hot Encoding

Creates binary columns.

Advantages:

- No artificial ordering
- Preferred for nominal categories

Widely used in production machine learning systems.



# 6. Introduction to Pipelines

Pipelines automate preprocessing and modeling.

Workflow:

Raw Data
→ Preprocessing
→ Model
→ Prediction


```

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression())
])

print(pipeline)
```


## Why Pipelines Matter

Benefits:

- Cleaner code
- Reproducibility
- Prevents data leakage
- Easier hyperparameter tuning



# Data Leakage

Data leakage occurs when information from the test set influences training.

This creates unrealistically high accuracy.

Pipelines help avoid leakage by applying transformations correctly.



# ColumnTransformer

Many datasets contain:

- Numerical variables
- Categorical variables

Different preprocessing techniques are required for each type.


```

from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), [0]),
        ('cat', OneHotEncoder(), [1])
    ]
)

print(preprocessor)
```


## Why ColumnTransformer?

Allows different transformations on different columns.

This is a standard approach in modern machine learning workflows.



# 7. Feature Engineering

Feature Engineering creates better representations of data.

Examples:

- Derived variables
- Polynomial features
- Aggregations
- Domain knowledge features


```

from sklearn.preprocessing import PolynomialFeatures
import numpy as np

X = np.array([[2],[3],[4]])

poly = PolynomialFeatures(degree=2)

print(poly.fit_transform(X))
```


## Polynomial Features

Original:

x

Expanded:

1, x, x²

This allows linear models to learn nonlinear relationships.



# Why Feature Engineering Works

Algorithms learn patterns from features.

Better features often improve performance more than switching algorithms.
