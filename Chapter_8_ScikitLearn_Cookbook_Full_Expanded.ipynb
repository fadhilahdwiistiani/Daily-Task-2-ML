# 1. Introduction to Decision Trees

Decision Tree adalah algoritma yang membagi data menjadi beberapa cabang berdasarkan aturan tertentu.

Karakteristik:

- Mudah dipahami
- Mudah divisualisasikan
- Tidak memerlukan scaling
- Dapat digunakan untuk klasifikasi maupun regresi


```

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(
    iris.data,
    iris.target,
    random_state=42
)

tree = DecisionTreeClassifier(
    random_state=42
)

tree.fit(X_train, y_train)

print("Accuracy:", tree.score(X_test,y_test))
```


## Code Explanation

DecisionTreeClassifier()

Membangun model pohon keputusan.

fit()

Melatih model menggunakan data training.

score()

Mengukur akurasi model.



# How Decision Trees Work

Langkah:

1. Memilih fitur terbaik
2. Membagi data menjadi subset
3. Mengulangi proses hingga kondisi berhenti tercapai

Tujuan:

Memaksimalkan pemisahan kelas.



# Splitting Criteria

Kriteria umum:

- Gini Impurity
- Entropy
- Information Gain

Semakin kecil impurity, semakin baik pemisahan data.



# Advantages of Decision Trees

- Interpretasi mudah
- Tidak memerlukan normalisasi
- Menangani data numerik dan kategorikal



# Limitations of Decision Trees

- Mudah overfitting
- Sensitif terhadap perubahan data kecil
- Variansi tinggi



# 2. Random Forests and Bagging

Random Forest merupakan kumpulan banyak decision tree.



## Bagging Concept

Bagging = Bootstrap Aggregating

Tujuan:

Mengurangi variansi model dengan membangun banyak model pada subset data berbeda.


```

from sklearn.ensemble import RandomForestClassifier

forest = RandomForestClassifier(
    n_estimators=100,
    random_state=42
)

forest.fit(X_train,y_train)

print("Accuracy:", forest.score(X_test,y_test))
```


## Why Random Forest Works

Random Forest:

- Menggunakan banyak tree
- Voting mayoritas
- Mengurangi overfitting
- Lebih stabil dibanding satu decision tree



# Feature Importance

Random Forest dapat menghitung tingkat kepentingan fitur.


```

for idx, value in enumerate(forest.feature_importances_):
    print(idx, round(value,4))
```


Feature Importance membantu memahami fitur mana yang paling berpengaruh terhadap prediksi.



# 3. Gradient Boosting Machines

Berbeda dengan Random Forest.

Random Forest:
- Tree independen

Gradient Boosting:
- Tree dibangun secara berurutan


```

from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(
    random_state=42
)

gb.fit(X_train,y_train)

print("Accuracy:", gb.score(X_test,y_test))
```


## How Gradient Boosting Works

Setiap tree baru berusaha memperbaiki kesalahan tree sebelumnya.

Hasil:

- Akurasi tinggi
- Generalisasi baik



# Comparing Random Forest and Gradient Boosting

| Random Forest | Gradient Boosting |
|--------------|------------------|
| Paralel | Sequential |
| Cepat | Lebih lambat |
| Lebih stabil | Lebih akurat |
| Variansi rendah | Bias rendah |



# Hyperparameter Tuning for Trees and Ensembles

Hyperparameter penting:

Decision Tree:
- max_depth
- min_samples_split
- min_samples_leaf

Random Forest:
- n_estimators
- max_depth

Gradient Boosting:
- learning_rate
- n_estimators
- max_depth


```

from sklearn.model_selection import GridSearchCV

param_grid = {
    'n_estimators':[50,100,200],
    'max_depth':[3,5,10]
}

grid = GridSearchCV(
    RandomForestClassifier(),
    param_grid,
    cv=5
)

grid.fit(X_train,y_train)

print(grid.best_params_)
print(grid.best_score_)
```


## Why Hyperparameter Tuning?

Karena parameter yang optimal berbeda untuk setiap dataset.



# Ensemble Methods

Ensemble Learning menggabungkan beberapa model.

Tujuan:

Meningkatkan performa dibanding model tunggal.



## Ensemble Categories

1. Bagging
2. Boosting
3. Stacking
4. Voting


```

from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

voting = VotingClassifier(
    estimators=[
        ('lr', LogisticRegression(max_iter=5000)),
        ('rf', RandomForestClassifier()),
        ('svm', SVC(probability=True))
    ],
    voting='soft'
)

voting.fit(X_train,y_train)

print(voting.score(X_test,y_test))
```
