# 1. Overview of Logistic Regression

Logistic Regression adalah algoritma supervised learning yang digunakan untuk klasifikasi.

Walaupun namanya regression, algoritma ini digunakan untuk memprediksi kelas.

Contoh:
- Spam vs Non-Spam
- Fraud vs Non-Fraud
- Disease vs Healthy



# Why Logistic Regression?

Keunggulan:

- Cepat
- Mudah diinterpretasikan
- Cocok untuk dataset kecil hingga menengah
- Menyediakan probabilitas prediksi


```

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

data = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    data.data,
    data.target,
    random_state=42
)

model = LogisticRegression(max_iter=5000)

model.fit(X_train, y_train)

print("Accuracy:", model.score(X_test, y_test))
```


## Code Explanation

fit()
→ melatih model

score()
→ menghitung accuracy

max_iter=5000
→ mencegah convergence warning



# Logistic Function (Sigmoid)

Logistic Regression menggunakan fungsi sigmoid:

p = 1/(1+e^-z)

Output selalu berada antara:

0 hingga 1

yang dapat diinterpretasikan sebagai probabilitas.



# How Logistic Regression Works

Workflow:

Input Features
↓
Linear Combination
↓
Sigmoid Function
↓
Probability
↓
Class Prediction



# 2. Multiclass Classification Techniques

Logistic Regression tidak hanya untuk binary classification.

Dapat digunakan untuk:

- Binary Classification
- Multiclass Classification


```

from sklearn.datasets import load_iris

iris = load_iris()

X = iris.data
y = iris.target

model = LogisticRegression(
    max_iter=5000,
    multi_class='auto'
)

model.fit(X,y)

print(model.score(X,y))
```


## Multiclass Strategies

1. One-vs-Rest (OvR)
2. Multinomial Logistic Regression

Scikit-learn dapat menangani keduanya secara otomatis.



# One-vs-Rest (OvR)

Misalnya terdapat 3 kelas:

A, B, C

Model akan membangun:

A vs Others

B vs Others

C vs Others

Kemudian memilih probabilitas tertinggi.



# Multinomial Logistic Regression

Pendekatan yang lebih langsung.

Semua kelas dipelajari secara bersamaan menggunakan softmax.


```

model = LogisticRegression(
    max_iter=5000,
    multi_class="multinomial"
)

model.fit(X,y)

print(model.score(X,y))
```


# 3. Regularization in Logistic Regression

Regularization membantu:

- Mengurangi overfitting
- Menstabilkan model
- Meningkatkan generalisasi



## Types of Regularization

L1 → Lasso

L2 → Ridge

ElasticNet → kombinasi keduanya


```

model_l1 = LogisticRegression(
    penalty='l1',
    solver='liblinear',
    max_iter=5000
)

model_l1.fit(X_train,y_train)

print(model_l1.score(X_test,y_test))
```

```

model_l2 = LogisticRegression(
    penalty='l2',
    max_iter=5000
)

model_l2.fit(X_train,y_train)

print(model_l2.score(X_test,y_test))
```


## Understanding C

Hyperparameter terpenting:

C

C kecil:
→ regularization kuat

C besar:
→ regularization lemah


```

for c in [0.01,0.1,1,10]:
    model = LogisticRegression(
        C=c,
        max_iter=5000
    )
    model.fit(X_train,y_train)
    print(c, model.score(X_test,y_test))
```


# 4. Multilabel Classification Concepts

Pada multilabel classification, satu data dapat memiliki lebih dari satu label.


```

from sklearn.datasets import make_multilabel_classification

X_ml, y_ml = make_multilabel_classification(
    n_samples=100,
    n_features=20,
    n_classes=3,
    random_state=42
)

print(X_ml.shape)
print(y_ml.shape)
```


Contoh:

Satu gambar dapat memiliki label:

- Car
- Road
- Traffic Sign

secara bersamaan.


```

from sklearn.multiclass import OneVsRestClassifier

multi_model = OneVsRestClassifier(
    LogisticRegression(max_iter=5000)
)

multi_model.fit(X_ml,y_ml)

print("Training Complete")
```


# 5. Model Evaluation Metrics

Evaluasi model klasifikasi tidak hanya menggunakan accuracy.



Metric umum:

- Accuracy
- Precision
- Recall
- F1 Score
- ROC-AUC
- Confusion Matrix


```

from sklearn.metrics import classification_report

pred = model.predict(X_test)

print(classification_report(y_test,pred))
```

```

from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,pred))
```


## Accuracy

Persentase prediksi benar.

Tidak selalu cukup untuk dataset tidak seimbang.



## Precision

Mengukur ketepatan prediksi positif.



## Recall

Mengukur kemampuan menemukan seluruh kasus positif.



## F1 Score

Harmonic mean antara precision dan recall.
