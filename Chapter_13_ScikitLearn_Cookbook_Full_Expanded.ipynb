# Getting a Model Ready for Deployment

Sebelum deployment:

- Model harus tervalidasi
- Hyperparameter sudah optimal
- Data preprocessing terdokumentasi
- Pipeline dapat direproduksi


```

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier

data = load_iris()

X = data.data
y = data.target

model = RandomForestClassifier(
    random_state=42
)

model.fit(X,y)

print("Model Ready")
```


# Serialization and Persistence Techniques

Model perlu disimpan agar dapat digunakan kembali tanpa retraining.


```

import joblib

joblib.dump(
    model,
    "rf_model.pkl"
)

print("Saved")
```


## Why Serialization?

Keuntungan:

- Tidak perlu melatih ulang model
- Deployment lebih cepat
- Konsistensi model terjaga


```

loaded_model = joblib.load(
    "rf_model.pkl"
)

print(
    loaded_model.predict(
        [X[0]]
    )
)
```


# Joblib vs Pickle

Joblib:
- Cepat untuk objek NumPy besar

Pickle:
- General purpose serialization

Joblib lebih sering digunakan pada model scikit-learn.

# Monitoring and Updating Deployed Models

Setelah deployment, model harus terus dipantau.



Metric yang dimonitor:

- Accuracy
- Latency
- Throughput
- Error Rate
- Data Drift


```

accuracy_history = [
    0.95,
    0.94,
    0.93,
    0.89
]

for idx, value in enumerate(accuracy_history):
    print(
        f"Week {idx+1}:",
        value
    )
```


Penurunan performa dapat mengindikasikan data drift atau concept drift.



# Data Drift

# Setting Up Deployment Pipelines

Deployment pipeline mengotomatisasi proses deployment.



Pipeline umum:

Code
↓
Testing
↓
Validation
↓
Packaging
↓
Deployment
↓
Monitoring


```

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    ("scaler", StandardScaler()),
    ("model", LogisticRegression())
])

pipeline.fit(X,y)

print("Pipeline Ready")
```


Pipeline memastikan preprocessing dan model selalu berjalan secara konsisten.



# Automating Deployment Checks

Sebelum deployment:

- Accuracy threshold
- Performance threshold
- Data validation


```

current_accuracy = 0.93
required_accuracy = 0.90

if current_accuracy >= required_accuracy:
    print("Deploy Model")
else:
    print("Reject Deployment")
```


Automated checks mengurangi risiko deployment model yang buruk.



# Practical Exercise 1

Saving and Reloading a Model Pipeline for Deployment


```

joblib.dump(
    pipeline,
    "deployment_pipeline.pkl"
)

loaded_pipeline = joblib.load(
    "deployment_pipeline.pkl"
)

print(
    loaded_pipeline.predict(
        [X[0]]
    )
)
```
