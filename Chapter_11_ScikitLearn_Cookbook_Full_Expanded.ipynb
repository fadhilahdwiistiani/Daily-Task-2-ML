Dataset training sudah mengandung outlier.

Model dilatih pada data normal, kemudian digunakan mendeteksi data baru yang tidak normal.

# 2. Understanding Isolation Forest

Isolation Forest merupakan salah satu algoritma anomaly detection paling populer.


```

from sklearn.ensemble import IsolationForest
from sklearn.datasets import make_blobs

X, _ = make_blobs(
    n_samples=300,
    centers=1,
    random_state=42
)

model = IsolationForest(
    contamination=0.05,
    random_state=42
)

labels = model.fit_predict(X)

print(labels[:20])
```


## Code Explanation

Output:

1  → Normal

-1 → Outlier

Isolation Forest bekerja dengan mengisolasi titik data menggunakan pohon acak.



# How Isolation Forest Works

Prinsip utama:

Outlier lebih mudah diisolasi dibanding data normal.

Semakin sedikit split yang diperlukan:

→ semakin besar kemungkinan data merupakan outlier.



# Advantages of Isolation Forest

- Cepat
- Skalabel
- Efektif pada data berdimensi tinggi
- Tidak memerlukan label



# 3. One-Class SVM for Novelty Detection

One-Class SVM digunakan untuk mempelajari batas data normal.


```

from sklearn.svm import OneClassSVM

svm_model = OneClassSVM(
    gamma='scale',
    nu=0.05
)

svm_model.fit(X)

pred = svm_model.predict(X)

print(pred[:20])
```


## One-Class SVM

Tujuan:

Mempelajari region yang berisi data normal.

Data di luar region dianggap anomali.



# Understanding Parameter nu

nu mengontrol:

- Proporsi outlier
- Kompleksitas boundary

Nilai besar:

Lebih banyak data dianggap anomali.



# 4. Detecting Outliers with LOF

LOF = Local Outlier Factor


```

from sklearn.neighbors import LocalOutlierFactor

lof = LocalOutlierFactor(
    n_neighbors=20
)

labels = lof.fit_predict(X)

print(labels[:20])
```


## How LOF Works

LOF membandingkan kepadatan lokal suatu titik dengan tetangganya.

Jika kepadatan jauh lebih rendah:

→ kemungkinan outlier.



# Advantages of LOF

- Cocok untuk data kompleks
- Memperhatikan struktur lokal
- Efektif pada cluster yang berbeda kepadatan



# 5. Evaluating Outlier Detection Models

Evaluasi anomaly detection lebih sulit dibanding klasifikasi biasa.



Metric umum:

- Precision
- Recall
- F1 Score
- ROC-AUC


```

from sklearn.metrics import classification_report
import numpy as np

y_true = np.ones(len(X))
y_true[:15] = -1

print(
    classification_report(
        y_true,
        labels
    )
)
```


Evaluasi dilakukan jika ground truth tersedia.
