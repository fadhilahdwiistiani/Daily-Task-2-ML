# 1. Introduction to Cross-Validation

Cross-validation adalah teknik evaluasi model dengan membagi dataset menjadi beberapa bagian (fold).

Tujuan:

- Mengukur performa model
- Mengurangi bias evaluasi
- Menghindari overfitting terhadap data test



# Why Cross-Validation Matters

Jika hanya menggunakan satu train-test split:

- Hasil dapat bergantung pada pembagian data
- Evaluasi menjadi kurang stabil

Cross-validation memberikan estimasi performa yang lebih reliabel.


```

from sklearn.datasets import load_breast_cancer
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

data = load_breast_cancer()

model = LogisticRegression(max_iter=5000)

scores = cross_val_score(
    model,
    data.data,
    data.target,
    cv=5
)

print(scores)
print(scores.mean())
```


## Code Explanation

cross_val_score()

Melatih dan mengevaluasi model pada beberapa fold secara otomatis.

cv=5

Menggunakan 5-fold cross-validation.



# How Cross-Validation Works

Contoh 5-Fold:

Fold 1 → Test
Fold 2-5 → Train

Kemudian diulang hingga setiap fold pernah menjadi test set.



# Advantages of Cross-Validation

- Estimasi lebih stabil
- Memanfaatkan seluruh data
- Cocok untuk dataset kecil



# 2. Advanced Cross-Validation Methods

Selain K-Fold terdapat metode lain.



## Types of Cross-Validation

1. K-Fold
2. Stratified K-Fold
3. Leave-One-Out
4. Repeated K-Fold
5. Group K-Fold


```

from sklearn.model_selection import StratifiedKFold

cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = cross_val_score(
    model,
    data.data,
    data.target,
    cv=cv
)

print(scores.mean())
```


## Stratified K-Fold

Menjaga proporsi kelas pada setiap fold.

Sangat penting untuk dataset tidak seimbang.



# Leave-One-Out Cross Validation (LOOCV)

LOOCV:

- Satu sampel menjadi test
- Sisanya menjadi training

Diulang untuk seluruh sampel.



# 3. Implementing Cross-Validation in scikit-learn

Scikit-learn menyediakan banyak utilitas untuk evaluasi model.


```

from sklearn.model_selection import KFold

cv = KFold(
    n_splits=5,
    shuffle=True,
    random_state=42
)

scores = cross_val_score(
    model,
    data.data,
    data.target,
    cv=cv
)

print(scores)
```


KFold digunakan untuk dataset umum tanpa kebutuhan stratifikasi.



# 4. Model Selection Techniques

Model Selection bertujuan memilih model terbaik dari beberapa kandidat.



Contoh model:

- Logistic Regression
- SVM
- Random Forest
- Gradient Boosting


```

from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC

models = {
    "Logistic": LogisticRegression(max_iter=5000),
    "RandomForest": RandomForestClassifier(),
    "SVM": SVC()
}

for name, model in models.items():
    scores = cross_val_score(
        model,
        data.data,
        data.target,
        cv=5
    )
    print(name, scores.mean())
```


Perbandingan ini membantu memilih model dengan performa terbaik.



# 5. Evaluating Model Generalizability

Generalisasi adalah kemampuan model bekerja pada data baru.



Model yang baik:

✓ Akurat pada training data

✓ Akurat pada unseen data


```

from sklearn.model_selection import learning_curve
import numpy as np

train_sizes, train_scores, test_scores = learning_curve(
    LogisticRegression(max_iter=5000),
    data.data,
    data.target,
    cv=5
)

print(train_sizes)
```


## Learning Curves

Digunakan untuk:

- Mendeteksi overfitting
- Mendeteksi underfitting
- Menilai kebutuhan data tambahan



# Validation Curves

Validation Curve membantu memahami pengaruh hyperparameter terhadap performa model.


```

from sklearn.model_selection import validation_curve

param_range = [0.01,0.1,1,10]

train_scores, test_scores = validation_curve(
    LogisticRegression(max_iter=5000),
    data.data,
    data.target,
    param_name="C",
    param_range=param_range,
    cv=5
)

print(test_scores.mean(axis=1))
```

# Practical Exercise 2

Tuning Hyperparameters with Grid Search


```

from sklearn.model_selection import GridSearchCV

param_grid = {
    "C":[0.01,0.1,1,10]
}

grid = GridSearchCV(
    LogisticRegression(max_iter=5000),
    param_grid,
    cv=5
)

grid.fit(data.data,data.target)

print(grid.best_params_)
```


Grid Search menggunakan cross-validation untuk menemukan parameter terbaik.
