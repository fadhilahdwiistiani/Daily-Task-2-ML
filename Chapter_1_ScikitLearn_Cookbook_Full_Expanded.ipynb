# 1. Introduction to scikit-learn Design Philosophy

scikit-learn was designed around four principles:

- Consistency
- Simplicity
- Modularity
- Reusability

Nearly every machine learning workflow follows:

1. Create estimator
2. Fit model
3. Predict or transform data

Core methods:

- fit()
- predict()
- transform()

This unified design makes switching between algorithms easy.



# Theory: Why API Consistency Matters

Without a standard API, every algorithm would require different commands.

scikit-learn solves this problem by ensuring:

```python
model.fit(X, y)
pred = model.predict(X_test)
```

works similarly across:

- Linear Regression
- Logistic Regression
- Random Forest
- SVM
- KNN
- Neural Networks

Benefits:

- Faster learning
- Easier experimentation
- Better maintainability
- Easier deployment



# 2. Understanding Estimators

Estimator is the central concept in scikit-learn.

An estimator is any object that learns patterns from data.

Examples:

- LinearRegression
- LogisticRegression
- RandomForestClassifier
- KMeans
- SVC

Main responsibilities:

- Learn from data
- Store learned parameters
- Produce predictions


````

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([1,2,3,3.5,5])

model = LinearRegression()
model.fit(X, y)

predictions = model.predict([[6],[7]])

print(predictions)
````


## Code Explanation

### model.fit(X,y)

Used to train the model.

The algorithm learns:

- slope (coefficient)
- intercept

### model.predict()

Generates predictions for unseen data.

Expected result:

```text
[5.75 6.70]
```

This demonstrates the estimator workflow.



# fit() vs predict()

### fit()

Learns patterns from training data.

### predict()

Applies learned patterns to new data.

Workflow:

Training Data
→ fit()
→ Trained Model
→ predict()
→ Predictions



# 3. fit_predict()

Some algorithms combine training and prediction into one step.

Mostly used in unsupervised learning.


````

from sklearn.cluster import KMeans
import numpy as np

X = np.array([[1],[2],[3],[4],[5]])

kmeans = KMeans(n_clusters=2, random_state=42)

labels = kmeans.fit_predict(X)

print(labels)
````


## Why fit_predict()?

K-Means does not predict target values.

Instead it:

1. Learns cluster centers
2. Assigns cluster labels

fit_predict() performs both tasks simultaneously.



# 4. Transformers and transform()

Transformers modify data before modeling.

Common transformers:

- StandardScaler
- MinMaxScaler
- OneHotEncoder
- PCA

Transformer workflow:

fit()
→ learn transformation

transform()
→ apply transformation


````

from sklearn.preprocessing import StandardScaler
import numpy as np

X = np.array([
    [1,2],
    [3,4],
    [5,6]
])

scaler = StandardScaler()

scaler.fit(X)

X_scaled = scaler.transform(X)

print(X_scaled)
````


## Code Explanation

StandardScaler computes:

- Mean
- Standard deviation

Formula:

z = (x - mean) / std

Result:

- Mean ≈ 0
- Standard deviation ≈ 1

Many ML algorithms perform better after scaling.



# fit_transform()

Shortcut method:

```python
scaler.fit_transform(X)
```

Equivalent to:

```python
scaler.fit(X)
scaler.transform(X)
```

Advantages:

- Cleaner code
- Faster preprocessing workflow


````

from sklearn.preprocessing import StandardScaler
import numpy as np

X = np.array([[1,2],[3,4],[5,6]])

scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled)
````


# 5. Custom Estimators and Transformers

One powerful feature of scikit-learn is extensibility.

Developers can create custom estimators by following the same API.

Requirements:

- fit()
- predict() OR transform()

Benefits:

- Reusable code
- Integration with pipelines
- Compatibility with GridSearchCV


````

from sklearn.base import BaseEstimator, TransformerMixin

class MultiplyByTwo(BaseEstimator, TransformerMixin):

    def fit(self, X, y=None):
        return self

    def transform(self, X):
        return X * 2
````


This custom transformer doubles all values while remaining compatible with scikit-learn pipelines.



# 6. Pipelines and Workflow Automation

Pipelines combine multiple steps into one object.

Example:

Raw Data
→ Scaling
→ Model
→ Prediction

Advantages:

- Cleaner workflow
- Prevents data leakage
- Easier deployment
- Easier hyperparameter tuning


````

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('model', LogisticRegression())
])

print(pipeline)
````


## Why Pipelines Matter

Without pipelines:

- Manual preprocessing
- Risk of mistakes

With pipelines:

```python
pipeline.fit(X_train, y_train)
pipeline.predict(X_test)
```

Everything happens automatically.

# 7. Common Attributes and Methods

After training, estimators store useful information.

Examples:

- coef_
- intercept_
- feature_importances_
- classes_


````

from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([[1],[2],[3],[4],[5]])
y = np.array([1,2,3,3.5,5])

model = LinearRegression()

model.fit(X,y)

print("Coefficient:", model.coef_)
print("Intercept:", model.intercept_)
print("R2:", model.score(X,y))
````


## Interpretation

coef_
→ slope

intercept_
→ y-axis intersection

score()
→ R² for regression

Higher R² generally means better fit.



# 8. Hyperparameter Tuning

Hyperparameters control model behavior before training.

Examples:

Random Forest:
- n_estimators
- max_depth

SVM:
- C
- gamma

KNN:
- n_neighbors


````

from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()

model.set_params(
    n_estimators=100,
    max_depth=10,
    random_state=42
)

print(model.get_params())
````


## Key Concepts

set_params()
→ modify hyperparameters

get_params()
→ inspect configuration

These methods are heavily used in:

- GridSearchCV
- RandomizedSearchCV



# Grid Search Theory

Grid Search tests combinations of hyperparameters.

Example:

n_estimators:
- 50
- 100
- 200

max_depth:
- 5
- 10
- 20

The algorithm evaluates all combinations and returns the best configuration.



# 9. Metadata and Tags

Modern versions of scikit-learn use metadata routing.

Metadata examples:

- sample_weight
- groups
- fit parameters

Purpose:

- Better pipeline communication
- Better cross-validation support
- Improved workflow flexibility



# Metadata Routing Concept

Router
→ passes metadata

Consumer
→ uses metadata

Example:

Sample Weights
→ Classifier

But

Sample Weights
→ ignored by StandardScaler

This improves efficiency.
