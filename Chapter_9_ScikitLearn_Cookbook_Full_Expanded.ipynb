# 2. Text Vectorization Techniques

Machine Learning tidak dapat langsung memahami teks.

Teks harus dikonversi menjadi angka.



Metode populer:

1. Bag of Words (BoW)
2. CountVectorizer
3. TF-IDF
4. N-Grams


```

from sklearn.feature_extraction.text import CountVectorizer

documents = [
    "machine learning is fun",
    "machine learning is powerful",
    "deep learning is powerful"
]

vectorizer = CountVectorizer()

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
print(X.toarray())
```


## CountVectorizer

Menghitung frekuensi kemunculan kata.

Output:

- Vocabulary
- Document-Term Matrix



# How CountVectorizer Works

Setiap kata unik menjadi fitur.

Contoh:

machine → feature

learning → feature

powerful → feature


```

from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer()

X_tfidf = tfidf.fit_transform(documents)

print(X_tfidf.toarray())
```


# TF-IDF

Term Frequency - Inverse Document Frequency

Tujuan:

Memberikan bobot lebih besar pada kata yang informatif dan mengurangi pengaruh kata yang terlalu umum.



# N-Grams

N-Gram mempertimbangkan urutan kata.

Contoh:

Unigram:
- machine
- learning

Bigram:
- machine learning


```

vectorizer = CountVectorizer(
    ngram_range=(1,2)
)

X = vectorizer.fit_transform(documents)

print(vectorizer.get_feature_names_out())
```


# 3. Feature Extraction from Text

# 4. Implementing Text Classification Models

Setelah teks diubah menjadi fitur numerik, model Machine Learning dapat digunakan.


```

from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB

pipeline = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('classifier', MultinomialNB())
])

texts = [
    "I love machine learning",
    "Machine learning is amazing",
    "I hate bugs",
    "Programming is frustrating"
]

labels = [1,1,0,0]

pipeline.fit(texts, labels)

prediction = pipeline.predict(
    ["Machine learning is powerful"]
)

print(prediction)
```


## Why Use Pipelines?

Pipeline menggabungkan:

- Vectorization
- Classification

ke dalam satu workflow.



# 5. Multiclass Classification Strategies

Banyak masalah NLP memiliki lebih dari dua kelas.

Contoh:

News Categories:

- Sports
- Politics
- Technology
- Entertainment



Strategi umum:

1. One-vs-Rest (OvR)
2. One-vs-One (OvO)
3. Native Multiclass Models


```

from sklearn.datasets import load_iris
from sklearn.linear_model import LogisticRegression

iris = load_iris()

model = LogisticRegression(
    max_iter=5000
)

model.fit(
    iris.data,
    iris.target
)

print(model.score(
    iris.data,
    iris.target
))
```


Logistic Regression mendukung multiclass classification secara langsung.



# 6. Evaluating Text Models

Evaluasi model NLP menggunakan metrik yang sama seperti klasifikasi lainnya.



Metric umum:

```

from sklearn.metrics import classification_report

pred = pipeline.predict(texts)

print(classification_report(labels,pred))
```


## Classification Report
