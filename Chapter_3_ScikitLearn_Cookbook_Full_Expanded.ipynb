# 1. Introduction to Dimensionality Reduction

Dimensionality Reduction adalah proses mengurangi jumlah fitur tanpa kehilangan informasi penting.

Tujuan utama:

- Mengurangi kompleksitas model
- Mengurangi noise
- Mempercepat training
- Mengurangi overfitting
- Mempermudah visualisasi



# Why Dimensionality Reduction is Essential

Masalah yang muncul ketika jumlah fitur terlalu banyak:

## Curse of Dimensionality

Semakin banyak dimensi:

- Data semakin sparse
- Jarak antar titik menjadi kurang bermakna
- Model membutuhkan lebih banyak data
- Training menjadi lebih lambat

Banyak algoritma Machine Learning mengalami penurunan performa ketika dimensi terlalu tinggi.



# Theoretical Foundations

Pendekatan utama:

1. Feature Selection
   - Memilih fitur terbaik

2. Feature Extraction
   - Membentuk fitur baru dari kombinasi fitur lama

PCA dan LDA termasuk Feature Extraction.



# 2. Transforming Datasets with PCA

Principal Component Analysis (PCA) adalah teknik reduksi dimensi paling populer.

Tujuan PCA:

- Menemukan arah variansi terbesar
- Mengurangi jumlah fitur
- Mempertahankan informasi penting


```

from sklearn.datasets import load_breast_cancer
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

data = load_breast_cancer()

X = data.data

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)

X_pca = pca.fit_transform(X_scaled)

print("Original Shape:", X.shape)
print("Reduced Shape :", X_pca.shape)
```


## Code Explanation

### StandardScaler()

PCA sensitif terhadap skala fitur sehingga data harus distandarisasi.

### PCA(n_components=2)

Mengurangi fitur menjadi 2 principal components.

### fit_transform()

Mempelajari struktur data dan menghasilkan representasi baru.



# Explained Variance

PCA menghasilkan informasi mengenai seberapa banyak variansi yang berhasil dipertahankan.


```

print(pca.explained_variance_ratio_)
print(sum(pca.explained_variance_ratio_))
```


Semakin besar explained variance ratio, semakin banyak informasi yang dipertahankan.



# How PCA Works

Langkah-langkah:

1. Standardisasi data
2. Hitung covariance matrix
3. Hitung eigenvalues
4. Hitung eigenvectors
5. Pilih komponen terbaik
6. Transformasikan data

Konsep matematika utama:

- Covariance Matrix
- Eigenvalues
- Eigenvectors



# PCA Visualization


```

import matplotlib.pyplot as plt

plt.scatter(
    X_pca[:,0],
    X_pca[:,1],
    alpha=0.7
)

plt.title("PCA Projection")
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.show()
```


# Hands-On Exercise: PCA

# 3. Maximizing Class Separability with LDA

Linear Discriminant Analysis (LDA) berbeda dari PCA.

PCA:
- Tidak menggunakan label

LDA:
- Menggunakan label kelas
- Memaksimalkan pemisahan antar kelas


```

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

y = data.target

lda = LinearDiscriminantAnalysis(
    n_components=1
)

X_lda = lda.fit_transform(X, y)

print(X_lda.shape)
```


## Why LDA?

LDA sangat efektif untuk:

- Classification
- Feature Extraction
- Reduksi dimensi pada data berlabel



# How LDA Works

Tujuan:

Memaksimalkan:

Between-Class Variance

dan

Meminimalkan:

Within-Class Variance

Sehingga kelas menjadi lebih mudah dipisahkan.



# PCA vs LDA

| PCA | LDA |
|------|------|
| Unsupervised | Supervised |
| Tidak memakai label | Memakai label |
| Maksimalkan variansi | Maksimalkan separasi kelas |
| Cocok visualisasi | Cocok klasifikasi |



# 4. Differences Between PCA and LDA

PCA fokus pada struktur data.

LDA fokus pada pemisahan kelas.

Jika tujuan:

Visualisasi → PCA

Klasifikasi → LDA



# 5. t-SNE and Data Visualization

t-SNE (t-Distributed Stochastic Neighbor Embedding)

digunakan untuk visualisasi data berdimensi tinggi.


```

from sklearn.manifold import TSNE

tsne = TSNE(
    n_components=2,
    random_state=42
)

X_tsne = tsne.fit_transform(X_scaled)

print(X_tsne.shape)
```


## Code Explanation

TSNE tidak digunakan untuk training model.

Tujuan utama:

- Visualisasi
- Exploratory Data Analysis
- Cluster inspection


```

plt.scatter(
    X_tsne[:,0],
    X_tsne[:,1],
    c=y,
    alpha=0.7
)

plt.title("t-SNE Visualization")
plt.show()
```


# How t-SNE Works

t-SNE:

1. Menghitung kedekatan titik pada ruang asli
2. Memetakan ke ruang berdimensi rendah
3. Mempertahankan struktur lokal

Kelebihan:

- Sangat bagus untuk visualisasi

Kekurangan:

- Lambat
- Tidak cocok untuk deployment



# Guidelines for Selecting Dimensionality Reduction Techniques

Gunakan:

### PCA
Jika:
- Banyak fitur numerik
- Ingin mengurangi noise
- Ingin mempercepat training

### LDA
Jika:
- Memiliki label
- Fokus pada klasifikasi

### t-SNE
Jika:
- Tujuan visualisasi
- Eksplorasi dataset



# Practical Steps for Selection

1. Analisis jumlah fitur
2. Evaluasi performa model
3. Uji PCA
4. Uji LDA
5. Bandingkan hasil
6. Pilih metode terbaik



# Impact on Model Performance

Dimensionality Reduction dapat:

✓ Mengurangi overfitting

✓ Mempercepat training

✓ Mengurangi memory usage

✓ Mempermudah visualisasi

Namun:

✗ Terlalu banyak reduksi dapat menghilangkan informasi penting.



# Understanding the Trade-Offs

Semakin sedikit dimensi:

+ Training lebih cepat

- Informasi bisa hilang

Tujuan:

Menemukan keseimbangan terbaik.



# Practical Exercise 1: PCA + Logistic Regression


```

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(
    X_pca,
    y,
    random_state=42
)

model = LogisticRegression(max_iter=5000)

model.fit(X_train, y_train)

print(model.score(X_test, y_test))
```


Tujuan:

Melihat apakah PCA masih mempertahankan informasi yang cukup untuk klasifikasi.
