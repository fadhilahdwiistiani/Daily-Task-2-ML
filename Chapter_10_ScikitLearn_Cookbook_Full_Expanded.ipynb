# 2. K-Means Clustering

K-Means merupakan algoritma clustering paling populer.


```

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X, y = make_blobs(
    n_samples=300,
    centers=3,
    random_state=42
)

kmeans = KMeans(
    n_clusters=3,
    random_state=42
)

labels = kmeans.fit_predict(X)

print(labels[:10])
```


## Code Explanation

n_clusters=3

Menentukan jumlah cluster yang ingin dibentuk.

fit_predict()

Melatih model dan menghasilkan label cluster.



# How K-Means Works

Langkah:

1. Pilih centroid awal
2. Hitung jarak ke centroid
3. Assign data ke centroid terdekat
4. Update centroid
5. Ulangi hingga konvergen



# Choosing K

Menentukan jumlah cluster merupakan tantangan utama K-Means.


```

inertia = []

for k in range(1,11):
    model = KMeans(
        n_clusters=k,
        random_state=42
    )
    model.fit(X)
    inertia.append(model.inertia_)

print(inertia)
```


## Elbow Method

Elbow Method digunakan untuk memilih jumlah cluster optimal.

Titik siku (elbow) menunjukkan nilai K terbaik.



# 3. Hierarchical Clustering

Hierarchical Clustering membangun struktur cluster berbentuk pohon.


```

from sklearn.cluster import AgglomerativeClustering

hier = AgglomerativeClustering(
    n_clusters=3
)

labels_hier = hier.fit_predict(X)

print(labels_hier[:10])
```


## Agglomerative Clustering

Pendekatan bottom-up:

- Setiap titik sebagai cluster
- Cluster digabung bertahap
- Hingga jumlah cluster tercapai



# Dendrogram Concept

Dendrogram digunakan untuk memvisualisasikan proses penggabungan cluster.



# Advantages of Hierarchical Clustering

- Tidak perlu menentukan K sejak awal
- Mudah divisualisasikan
- Cocok untuk analisis eksploratif



# 4. Density-Based Clustering with DBSCAN

DBSCAN menggunakan konsep kepadatan data.


```

from sklearn.cluster import DBSCAN

dbscan = DBSCAN(
    eps=0.5,
    min_samples=5
)

labels_db = dbscan.fit_predict(X)

print(labels_db[:10])
```


## Why DBSCAN?

DBSCAN mampu:

- Menemukan cluster berbentuk arbitrer
- Menangani noise
- Tidak memerlukan jumlah cluster di awal



# Understanding Parameters

eps

Radius pencarian tetangga.

min_samples

Jumlah minimum tetangga untuk membentuk cluster.



# K-Means vs DBSCAN

| K-Means | DBSCAN |
|----------|----------|
| Perlu K | Tidak perlu K |
| Sensitif outlier | Tahan outlier |
| Cluster bulat | Cluster arbitrer |



# 5. Cluster Evaluation Metrics

Clustering tidak memiliki label sehingga evaluasinya berbeda.


```

from sklearn.metrics import silhouette_score

score = silhouette_score(X, labels)

print(score)
```


## Silhouette Score

Rentang:

-1 hingga 1

Semakin tinggi → cluster semakin baik.



# Additional Metrics

- Silhouette Score
- Davies-Bouldin Index
- Calinski-Harabasz Score


```

from sklearn.metrics import davies_bouldin_score

print(
    davies_bouldin_score(
        X,
        labels
    )
)
```


# 6. Choosing the Right Clustering Algorithm

# 7. Advanced Clustering Techniques

Metode lanjutan:

- Gaussian Mixture Models (GMM)
- Spectral Clustering
- Mean Shift
- Birch


```

from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(
    n_components=3,
    random_state=42
)

gmm.fit(X)

labels_gmm = gmm.predict(X)

print(labels_gmm[:10])
```


## Gaussian Mixture Models

Mengasumsikan data berasal dari beberapa distribusi Gaussian.



# Practical Exercise 1 – Clustering with K-Means on Iris Dataset


```

from sklearn.datasets import load_iris

iris = load_iris()

kmeans_iris = KMeans(
    n_clusters=3,
    random_state=42
)

kmeans_iris.fit(iris.data)

print(kmeans_iris.labels_[:10])
```


# Practical Exercise 2 – Comparing DBSCAN and K-Means on Moon Data


```

from sklearn.datasets import make_moons

X_moon, _ = make_moons(
    n_samples=300,
    noise=0.05,
    random_state=42
)

db = DBSCAN(eps=0.2)

db.fit(X_moon)

print(set(db.labels_))
```


Moon dataset menunjukkan keunggulan DBSCAN dibanding K-Means pada bentuk cluster non-linear.



# Practical Exercise 3 – Clustering High-Dimensional Data with PCA and GMMs


```

from sklearn.decomposition import PCA

pca = PCA(n_components=2)

X_reduced = pca.fit_transform(iris.data)

gmm.fit(X_reduced)

print("Done")
```


PCA membantu mengurangi dimensi sebelum clustering dilakukan.
