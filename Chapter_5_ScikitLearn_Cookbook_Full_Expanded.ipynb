# Introduction to Linear Models

Linear models merupakan fondasi banyak algoritma Machine Learning. Ide utamanya adalah memodelkan hubungan linear antara fitur dan target.

# Linear Regression

Linear Regression digunakan untuk memprediksi nilai kontinu berdasarkan kombinasi linear fitur.

```

from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression

X, y = make_regression(
    n_samples=500,
    n_features=5,
    noise=10,
    random_state=42
)

model = LinearRegression()
model.fit(X,y)

print("R² Score:", model.score(X,y))
print("Coefficients:", model.coef_)
```


## Code Explanation

### LinearRegression()

Membuat model regresi linear.

### fit()

Mempelajari hubungan antara fitur dan target.

### coef_

Menampilkan kontribusi masing-masing fitur.



# How Linear Models Work

Persamaan umum:

y = β₀ + β₁x₁ + β₂x₂ + ... + βₙxₙ

Model berusaha mencari koefisien terbaik untuk meminimalkan error prediksi.



# Ridge and Lasso Regression

Masalah utama Linear Regression:

- Overfitting
- Sensitif terhadap multikolinearitas

Solusi:

- Ridge Regression
- Lasso Regression


```

from sklearn.linear_model import Ridge

ridge = Ridge(alpha=1.0)
ridge.fit(X,y)

print("Ridge Score:", ridge.score(X,y))
```


## Ridge Regression

Ridge menambahkan penalti L2.

Tujuan:

Mengurangi nilai koefisien yang terlalu besar sehingga model menjadi lebih stabil.


```

from sklearn.linear_model import Lasso

lasso = Lasso(alpha=0.1)
lasso.fit(X,y)

print("Lasso Score:", lasso.score(X,y))
print("Non-zero Coefficients:", sum(lasso.coef_ != 0))
```


## Lasso Regression

Lasso menggunakan regularisasi L1.

Keunggulan:

Selain mengurangi overfitting, Lasso juga dapat melakukan feature selection secara otomatis.



# Difference Between Ridge and Lasso

| Ridge | Lasso |
|---------|---------|
| L2 Penalty | L1 Penalty |
| Semua fitur dipertahankan | Dapat menghilangkan fitur |
| Cocok untuk multikolinearitas | Cocok untuk feature selection |



# ElasticNet and Regularization

ElasticNet menggabungkan Ridge dan Lasso.


```

from sklearn.linear_model import ElasticNet

elastic = ElasticNet(
    alpha=0.1,
    l1_ratio=0.5
)

elastic.fit(X,y)

print("ElasticNet Score:", elastic.score(X,y))
```


## Why ElasticNet?

ElasticNet mengombinasikan:

- L1 Regularization
- L2 Regularization

sehingga sering memberikan performa yang lebih stabil pada dataset kompleks.

# Understanding Alpha

Hyperparameter terpenting:

alpha

Nilai alpha besar:

- Regularization kuat
- Koefisien mengecil

Nilai alpha kecil:

- Model lebih fleksibel


```

for alpha in [0.01,0.1,1,10]:
    model = Ridge(alpha=alpha)
    model.fit(X,y)
    print(alpha, model.score(X,y))
```


Eksperimen ini menunjukkan bagaimana perubahan alpha memengaruhi performa model.



# Regression and Regularization

Workflow umum:

1. Siapkan data
2. Split train-test
3. Pilih model
4. Terapkan regularization
5. Evaluasi performa
6. Tuning hyperparameter


```

from sklearn.model_selection import GridSearchCV

param_grid = {
    "alpha":[0.001,0.01,0.1,1,10]
}

grid = GridSearchCV(
    Ridge(),
    param_grid,
    cv=5
)

grid.fit(X,y)

print(grid.best_params_)
print(grid.best_score_)
```


# Hyperparameter Tuning

GridSearchCV digunakan untuk menemukan nilai alpha terbaik secara otomatis.
