# Maximum Margin Principle

Pembahasan teori mendalam mengenai maximum margin principle dan perannya dalam klasifikasi modern menggunakan Support Vector Machines.

```

from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC

data = load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(
    data.data,
    data.target,
    random_state=42
)

svm = SVC()
svm.fit(X_train, y_train)

print("Accuracy:", svm.score(X_test, y_test))
```


## Code Explanation

`SVC()` adalah implementasi Support Vector Classifier dalam scikit-learn.

Tahapan:

1. Membuat model SVM
2. Melatih menggunakan `fit()`
3. Mengevaluasi menggunakan `score()`



# Maximum Margin Principle

SVM berusaha menemukan hyperplane terbaik yang memisahkan kelas dengan margin terbesar.

Semakin besar margin:

- Generalisasi lebih baik
- Risiko overfitting lebih kecil



# Kernel Functions and Their Applications

Kernel memungkinkan SVM menangani data non-linear.

Kernel umum:

- Linear
- Polynomial
- RBF (Radial Basis Function)
- Sigmoid


```

from sklearn.svm import SVC

linear_svm = SVC(kernel='linear')
linear_svm.fit(X_train, y_train)

print(linear_svm.score(X_test, y_test))
```


## Linear Kernel

Cocok ketika data relatif linear dan jumlah fitur besar.


```

rbf_svm = SVC(
    kernel='rbf',
    gamma='scale'
)

rbf_svm.fit(X_train, y_train)

print(rbf_svm.score(X_test, y_test))
```


## RBF Kernel

Kernel paling populer.

Mampu membentuk decision boundary kompleks untuk memisahkan data non-linear.


```

poly_svm = SVC(
    kernel='poly',
    degree=3
)

poly_svm.fit(X_train, y_train)

print(poly_svm.score(X_test, y_test))
```


## Polynomial Kernel

Menambahkan kemampuan mempelajari hubungan polinomial antar fitur.



# Tuning SVM Parameters

Hyperparameter penting:

- C
- gamma
- kernel
- degree


```

from sklearn.model_selection import GridSearchCV

param_grid = {
    'C':[0.1,1,10],
    'gamma':['scale',0.01,0.001],
    'kernel':['rbf']
}

grid = GridSearchCV(
    SVC(),
    param_grid,
    cv=5
)

grid.fit(X_train,y_train)

print(grid.best_params_)
print(grid.best_score_)
```


## Understanding C

C kecil:

- Margin lebih besar
- Regularisasi lebih kuat

C besar:

- Margin lebih kecil
- Model lebih kompleks



## Understanding Gamma

Gamma menentukan seberapa jauh pengaruh satu sampel terhadap decision boundary.

Gamma tinggi:
- Boundary kompleks

Gamma rendah:
- Boundary lebih halus
